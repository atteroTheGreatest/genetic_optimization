```python

from __future__ import division
import numpy as np


t = 0.1
def function(x):
    return 4 * np.sign(np.sin(x) * np.sin(t)) + np.sin(5 * x)

with plt.style.context(('ggplot')):
    xs = np.linspace(0, 10, 200)
    plt.figure(figsize=(10, 6))
    plt.plot(xs, function(xs), linewidth=5)
    plt.xlabel('x', size=20)
    plt.ylabel('fitness', size=20)
```

```python
import random
from deap import base
from deap import tools
from deap import creator

from collections import Counter

toolbox = base.Toolbox()


def to_dec(genotype):
    number = 0
    base = 2
    for i, x in enumerate(genotype):
        number += int(x) * base ** i
    return number


def evaluateInd(individual):
    # Do some computation
    number = to_dec(individual)
    return function(number / 20.0) + 5.2,

toolbox.register("mate", tools.cxTwoPoint)
toolbox.register("mutate", tools.mutFlipBit, indpb=0.1)
toolbox.register("select", tools.selTournament, tournsize=3)
toolbox.register("evaluate", evaluateInd)

#CXPB, MUTPB, NGEN = 0.5, 0.3, 10000

CXPB, MUTPB, NGEN = 0.80, 0.1, 1000

from functools import partial
import matplotlib.gridspec as gridspec

gs = gridspec.GridSpec(3, 4)
fig = plt.figure(figsize=(12, 8))


IND_SIZE = 7
toolbox.register("attribute", partial(random.choice, [0, 1]))

creator.create("FitnessMax", base.Fitness, weights=(1.0,))
creator.create("Individual", list, fitness=creator.FitnessMax)
toolbox.register("individual", tools.initRepeat, creator.Individual,
                 toolbox.attribute, n=IND_SIZE)

toolbox.register("population", tools.initRepeat, list, toolbox.individual)

pop = toolbox.population(n=200)

t = 0

ts = []
fits = []

for g in range(NGEN):
    t += 0.01
    # Select the next generation individuals
    offspring = toolbox.select(pop, len(pop))
    # Clone the selected individuals
    offspring = map(toolbox.clone, offspring)

    # Apply crossover on the offspring
    for child1, child2 in zip(offspring[::2], offspring[1::2]):
        if random.random() < CXPB:
            toolbox.mate(child1, child2)
            del child1.fitness.values
            del child2.fitness.values

    # Apply mutation on the offspring
    for mutant in offspring:
        if random.random() < MUTPB:
            toolbox.mutate(mutant)
            del mutant.fitness.values

    # Evaluate the individuals with an invalid fitness
    invalid_ind = [ind for ind in offspring if not ind.fitness.valid]
    fitnesses = toolbox.map(toolbox.evaluate, invalid_ind)
    
    #print(fitnesses)
    #print(Counter([tuple(x) for x in pop]).most_common(5))
    for ind, fit in zip(invalid_ind, fitnesses):
        ind.fitness.values = fit

    # The population is entirely replaced by the offspring
    pop[:] = offspring
    
    if g % 45 == 0:
        xs = [to_dec(individual[0]) / 20 for individual in Counter([tuple(x) for x in pop]).most_common(3)]
        ss = [individual[1] * 10 for individual in Counter([tuple(x) for x in pop]).most_common(3)]
        ys = [function(x) for x in xs]
        ts.append(t)
        fits.append(np.average(fitnesses))
        
        if g % 90 == 0:
            ax1 = plt.subplot(gs[g // 90])
            ax1.scatter(xs, ys, s=ss)
            zs = np.linspace(0, 10, 200)
            ax1.plot(zs, function(zs))
            ax1.set_title('time=%s' % t)
            ax1.xlabel = 'x'
            ax1.ylabel = 'fitness'
        

fig.suptitle('Best individuals', fontsize=20)
```

```python
with plt.style.context(('ggplot')):
    plt.figure(figsize=(10, 6))
    plt.ylim(0, 1.3)
    plt.xlim(0, 10)
    t = 0
    ts_m = []
    maxes = []
    plt.plot(ts, np.array(fits) / max(fits), 'o', markersize=15, label="Normalized medium fitness")
    print(np.average(fits))
    for x in range(200):
        t += 0.05

        xs = np.linspace(0, 10, 1000)
        ys = np.array([function(x) for x in xs])
        big = np.max(ys)
        mask = ys >= big - 0.01
        ts_m.append(t)
        maxes.append(xs[mask][0])

    plt.plot(ts_m, np.array(maxes) / max(maxes), linewidth=4, label="Normalized xs of first maximum")
    plt.xlabel('time', size=20)
    plt.ylabel('fitness, x', size=15)
    plt.legend(loc=0)
```

```python
from matplotlib import cm
```

```python
from numpy import linspace, pi, meshgrid

phi_m = linspace(0, 10, 100)
phi_p = linspace(0, 10, 100)
X,Y = meshgrid(phi_p, phi_m)

def function_by_t(x, t):
    return 4 * np.sign(np.sin(x) * np.sin(t)) + np.sin(5 * x)

Z = function_by_t(X, Y).T
```

```python
with plt.style.context('bmh'):
    fig, ax = plt.subplots(figsize=(8, 6))
    
    p = ax.pcolor(X, Y, Z, cmap=cm.RdBu, vmin=Z.min(), vmax=abs(Z).max())
    plt.xlabel('time', size=20)
    plt.ylabel('x', size=20)
    cb = fig.colorbar(p, ax=ax)
```
